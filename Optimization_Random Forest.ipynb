```python
import numpy as np
import scipy.io as sio
from pymoo.core.problem import Problem
from pymoo.algorithms.moo.nsga3 import NSGA3
from pymoo.optimize import minimize
from pymoo.visualization.scatter import Scatter
from pymoo.util.reference_direction import UniformReferenceDirectionFactory

# Step 1: Load the data from the .mat file
data = sio.loadmat('training_data.mat')
X_train = data['X_train']  # Features
Y_train = data['Y_train']  # Targets

from sklearn.ensemble import RandomForestRegressor

# Step 2: Train a Random Forest regressor model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, Y_train)

import matplotlib.pyplot as plt

# Calculate predictions from the Random Forest model
Y_rf_preds = rf_model.predict(X_train)

# Calculate correlation coefficients (R) between predicted and actual values for each output variable
correlation_coeffs = []
for i in range(Y_train.shape[1]):
    correlation_coeff = np.corrcoef(Y_train[:, i], Y_rf_preds[:, i])[0, 1]
    correlation_coeffs.append(correlation_coeff)

# Plot the accuracy of Random Forest with respect to each output variable
labels = ['Purity', 'Recovery', 'Productivity', 'Energy Requirement', 'TCR']
plt.figure(figsize=(10, 6))
plt.bar(labels, correlation_coeffs, color='skyblue')
plt.xlabel('Output Variable')
plt.ylabel('Correlation Coefficient (R)')
plt.title('Accuracy of Random Forest Model for Each Output Variable')
plt.ylim(0, 1)
plt.grid(axis='y')
plt.show()


```

```python
class MLDTLZ(Problem):
    def __init__(self, model, **kwargs):
        super().__init__(n_var=8, n_obj=5, n_constr=0, xl=[1, 1e5, 0.1e5, 10, 0.1, 0.01, 0, 1], 
                         xu=[7, 10e5, 0.5e5, 1000, 2, 0.99, 1, 16], **kwargs)
        self.model = model

    def _evaluate(self, X, out, *args, **kwargs):
        # Call the trained model to make predictions
        Y_pred = self.model.predict(X)

        purity = -Y_pred[:, 0]
        recovery = -Y_pred[:, 1]
        productivity = -Y_pred[:, 2]
        energy_requirement = Y_pred[:, 3]  
        TCR = Y_pred[:, 4] 

        # Assign objectives to the output dictionary
        out["F"] = np.column_stack([purity, recovery, productivity, energy_requirement, TCR])

        # Calculate constraints
        #constraint1 = 0.6 - purity  # Purity constraint: Should be positive
        #constraint2 = 0.7 - recovery  # Recovery constraint: Should be positive
        #constraint3 = -energy_requirement  # Productivity constraint: Should be non-negative
        #constraint4 = -TCR
        
        # Assign constraints to the output dictionary
        #out["G"] = np.column_stack([constraint1, constraint2])




# Step 4: Create the MLDTLZ problem instance with Random Forest model
problem = MLDTLZ(rf_model)

# Step 5: Define the number of reference directions
n_obj = problem.n_obj

# Step 6: Create the reference directions using UniformReferenceDirectionFactory
#ref_dirs = UniformReferenceDirectionFactory(n_obj, n_points=70).do()
ref_dirs = UniformReferenceDirectionFactory(n_obj, n_points=126).do()


# Step 7: Initialize the NSGA3 algorithm with reference directions and population size
algorithm = NSGA3(ref_dirs=ref_dirs, pop_size=126)



# Step 8: Execute the optimization with reduced maximum number of generations
res = minimize(problem,
               algorithm,
               seed=1,
               termination=('n_gen', 600))  # Reduced to 300 generations


import matplotlib.pyplot as plt

# Extract the objectives from the optimization result
objectives = res.F

# Extract purity and recovery
purity_recovery = objectives[:, :2]

# Extract energy and productivity
energy_productivity = objectives[:, [3, 2]]  # Energy is at index 3, productivity is at index 2

# Convert purity and recovery into positive values
purity_recovery[:, 0] = np.abs(purity_recovery[:, 0])
purity_recovery[:, 1] = np.abs(purity_recovery[:, 1])

# Convert energy and productivity into positive values
energy_productivity[:, 0] = np.abs(energy_productivity[:, 0])
energy_productivity[:, 1] = np.abs(energy_productivity[:, 1])

# Plot Pareto front between purity and recovery
plt.figure(figsize=(8, 6))
plt.scatter(purity_recovery[:, 0], purity_recovery[:, 1], color="blue", s=30)
plt.xlabel("Purity")
plt.ylabel("Recovery")
plt.title("Pareto Front between Purity and Recovery")
plt.grid(True)
plt.tight_layout()

# Save the plot as a PNG file
plt.savefig("pareto_front_purity_recovery.png")

# Show the plot
plt.show()

# Plot Pareto front between energy and productivity
plt.figure(figsize=(8, 6))
plt.scatter(energy_productivity[:, 0], energy_productivity[:, 1], color="red", s=30)
plt.xlabel("Energy")
plt.ylabel("Productivity")
plt.title("Pareto Front between Energy and Productivity")
plt.grid(True)
plt.tight_layout()

# Save the plot as a PNG file
plt.savefig("pareto_front_energy_productivity.png")

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import itertools

# Convert purity, recovery, and productivity into positive values
positive_objectives = res.F.copy()
positive_objectives[:, :3] = np.abs(positive_objectives[:, :3])

# Define the output names
output_names = ['Purity', 'Recovery', 'Energy', 'Productivity', 'Total Cost']

# Create a grid of plots
fig, axes = plt.subplots(5, 5, figsize=(15, 15))

# Iterate over each pair of objectives and plot the Pareto front
for i, j in itertools.product(range(5), repeat=2):
    if i == j:
        # Plot the histogram for the diagonal
        axes[i, j].hist(positive_objectives[:, i], bins=20, color="skyblue", edgecolor="black", alpha=0.7)
        axes[i, j].set_xlabel(output_names[i])
        axes[i, j].set_ylabel("Frequency")
    else:
        # Plot the scatter plot for off-diagonal
        axes[i, j].scatter(positive_objectives[:, j], positive_objectives[:, i], color="blue", s=10)
        axes[i, j].set_xlabel(output_names[j])
        axes[i, j].set_ylabel(output_names[i])

# Adjust layout
plt.tight_layout()

# Save each plot as a PNG file
for i in range(5):
    for j in range(5):
        plt.figure()
        if i == j:
            plt.hist(positive_objectives[:, i], bins=20, color="skyblue", edgecolor="black", alpha=0.7)
            plt.xlabel(output_names[i])
            plt.ylabel("Frequency")
        else:
            plt.scatter(positive_objectives[:, j], positive_objectives[:, i], color="blue", s=10)
            plt.xlabel(output_names[j])
            plt.ylabel(output_names[i])
        plt.title("Pareto Front between " + output_names[i] + " and " + output_names[j])
        plt.grid(True)
        plt.tight_layout()
        plt.savefig("pareto_front_" + output_names[i].lower() + "_" + output_names[j].lower() + ".png")
        plt.close()

# Show the grid of plots
plt.show()
```
